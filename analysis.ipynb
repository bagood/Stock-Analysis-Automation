```python
import numpy as np
import pandas as pd
```

```python
median_gain_10dd_performances = pd.read_csv('database/modelPerformances/medianGain/10dd.csv')
median_gain_15dd_performances = pd.read_csv('database/modelPerformances/medianGain/15dd.csv')
max_loss_10dd_performances = pd.read_csv('database/modelPerformances/maxLoss/10dd.csv')
max_loss_15dd_performances = pd.read_csv('database/modelPerformances/maxLoss/15dd.csv')
```

```python
median_gain_10dd_forecasts = pd.read_csv('database/forecastedStocks/medianGain/10dd.csv')
median_gain_15dd_forecasts = pd.read_csv('database/forecastedStocks/medianGain/15dd.csv')
max_loss_10dd_forecasts = pd.read_csv('database/forecastedStocks/maxLoss/10dd.csv')
max_loss_15dd_forecasts = pd.read_csv('database/forecastedStocks/maxLoss/15dd.csv')
```

# 10 dd Analysis

```python
merged_win_data =  pd.merge(
                            median_gain_10dd_forecasts.sort_values('Forecast High Gain 10dd', ascending=False).head(20),
                            median_gain_10dd_performances,
                            on='Kode',
                            how='inner'
                        )

filtered_win_data = merged_win_data[merged_win_data['Threshold'] > 0] \
                        .sort_values('Threshold', ascending=False) \
                        .reset_index(drop=True)

selected_win_emiten = filtered_win_data['Kode'].values
```

```python
filtered_win_data
```

```python
merged_lose_data =  pd.merge(
                                max_loss_10dd_forecasts[max_loss_10dd_forecasts['Kode'].isin(selected_win_emiten)],
                                max_loss_10dd_performances,
                                on='Kode',
                                how='inner'
                            ) \
                        .sort_values('Threshold', ascending=False) \
                        .reset_index(drop=True)
```

```python
merged_lose_data
```
